# Exploring Sound Devices with `Sounddevice`

Prerequisites: install modules `sounddevice` and possibly `soundfile`



```python
import soundfile as sf
import sounddevice as sd
import pyaudio
import numpy as np
import time
import matplotlib.pyplot as plt
```

## Querying sound cards / sound devices



```python
# return a device list
devices = sd.query_devices()
len(devices)

# iterate over devices an display properties
for dev in devices:
    print(f"{dev}\n")
```

```python
# default input / output device
# select specific device
name = 'Mikrofon (Sound Blaster Play! 3'
microfones = [mic_dev for mic_dev in devices if mic_dev['name'] == name]
for microfone in microfones:
    print(f"{microfone}")
```

```python
# similarly the same device information may be obtained using the 'index'
microfone_device = sd.query_devices(33)
print(f"{microfone_device}")
```

---

## Recording 

Now that we know the device index for the microfone and the sampling rate we can record a short audio track.



```python
duration =5.0  # seconds
fs = int(microfone_device['default_samplerate'])
recording = sd.rec(int(duration * fs), samplerate=fs, channels=1)
print(f"recording started")
sd.wait()
print(f"recording finished")
print(f"len(recording): {len(recording)}")
```

```python
fname = './recordings/record_1.wav'

sf.write(fname, recording, fs)
```

```python
tvec = np.arange(len(recording)) / fs

fig1 = plt.figure(1, figsize=[8, 4])
ax_f1 = fig1.add_subplot(1, 1, 1)
ax_f1.plot(tvec, recording, linewidth=0.5, alpha=0.5,  color='k', label='audio record')
ax_f1.grid(True)
ax_f1.set_ylabel('y')
ax_f1.set_xlabel('time [s]')
ax_f1.legend()
ax_f1.set_title(f"audio recording: fs = {fs:10.2f} [Hz]");
```

## Capturing Audio using the Callback Interface / Version 1

The code example closely replicates an example from the documentation `python-sounddevice
Release 0.4.6-12-ga7c3259`

To avoid use of global parameters which can be accessed / modified in the callback the callback function is wrapped into 
a partial function using `functools`. 

A memory of numpy array to which audiosamples are copied is *preallocated* before any callback is invoked.

The signal activity is measured for each processed callback and the result is stored in an array. 

It should be emphasized that the callback not only collects audio samples into a queue but does some processing as well directly in the callback.
This may not be an optimum solution.

In a later version of the program, other approaches are tried with the aim, to offload such processing from the callback.


```python
import queue
from functools import partial

q = queue.Queue()

duration = 4.0 # seconds
nr_samples = int(duration * fs + fs)

update_rate = 50
nr_markers = int(duration * update_rate)
marker_samples = np.zeros(nr_markers, dtype=np.int32)

activity_samples = np.zeros(nr_markers, dtype=np.float32)
activity_samples[:] = np.NaN

# preallocation of memory (slightly larger than required)
audio_samples = np.zeros(nr_samples, dtype=np.float32)

# the callback shall update the starting position idx with every call
# this is not possible using idx directly as a positional argument because an integer is passed by value not
# by reference
# -> wrapping idx into a dictionary works because the dictionary is passed by reference
idx = 0
idx_dict = dict()
idx_dict['idx'] = idx
idx_dict['count'] = int(0)

# a less elegant approach is to use global parameters
# the callback (uses numpy interally)
def callback_global(indata, frames, time, status):
    global idx
    global audio_samples
    if status:
        print(status)
    q.put(indata.copy())
    data = q.get()
    audio_samples[idx:idx + len(data)] = data[:,0]
    idx = idx + len(data)

# the callback (uses numpy interally)
def callback_ref(idx_dict, q, audio_samples, marker_samples, activity_samples, indata, frames, time, status):
    idx = idx_dict['idx']
    count = idx_dict['count']
    if status:
        print(status)
    q.put(indata.copy())
    data = q.get()
    audio_samples[idx:idx + len(data)] = data[:,0]
    marker_samples[count] = idx
    activity_samples[count] = np.sum(np.abs(data[:,0]))
    idx_dict['idx'] = idx + len(data)
    idx_dict['count'] = count + 1

wrapped_cb = partial(callback_ref, idx_dict, q, audio_samples, marker_samples, activity_samples)
    
print("begin capturing data")
with sd.InputStream(samplerate=fs, device=1, channels=1, callback=wrapped_cb):
    sd.sleep(int(duration * 1000))

print("end capturing data")
```

```python
# check size of queue
# -> should be empty 
q.qsize()
```

## Capturing Audio using the Callback Interface / Version 2


In this version most of the processing shall be done external to the callback function.



```python
q = queue.Queue()

duration = 4.0 # seconds
nr_samples = int(duration * fs)

# preallocation of memory 
audio_samples = np.zeros(nr_samples, dtype=np.float32)

# the callback (uses numpy interally)
# -> the main purpose of the callback is to transfer audio samples to a queue
#    any further processing is done outside the callback function
def callback_ref2(q, indata, frames, time, status):
    if status:
        print(status)
    q.put(indata.copy())
    
wrapped_cb2 = partial(callback_ref2, q)
    
print("begin capturing data")
idx = 0
count_samples = 0
# the processing of audio data is done in the while loop thus freeing resources 
# from the callback function
with sd.InputStream(samplerate=fs, device=1, channels=1, callback=wrapped_cb2):
    while True:
        data = q.get()
        ndata = len(data)
        count_samples += ndata
        if count_samples >= nr_samples:
            break
        audio_samples[idx:idx + ndata] = data[:,0]
        idx = idx + ndata
        
print("end capturing data")
```

```python
fig3 = plt.figure(3, figsize=[8, 4])
ax_f31 = fig3.add_subplot(1, 1, 1)
ax_f31.plot(audio_samples, linewidth=0.5, alpha=0.5,  color='k', label='audio record')
ax_f31.grid(True)
ax_f31.set_ylabel('y')
ax_f31.set_xlabel('[n]')
ax_f31.legend()
ax_f31.set_title(f"audio recording: fs = {fs:10.2f} [Hz]");
```

## Capturing Audio using the Callback Interface / Version 3

Reusing the ideas from version 2 (postprocessing done separate from callback) the audio samples shall be collected in cyclically in a number of buffers.

**some properties of the implementation**

1) Each buffer shall hold about 10 seconds of audio samples

2) The number of buffers is chosen to be `5`.

3) A list is maintained into which the current number of audio samples of each buffer is written.

4) The collection of audio samples runs in an infinite `while-loop`. The while loop exits once a specific number of buffers have been updated.

5) The cpu-load parameter of the input stream is monitored. With cpu-load of the order of only a fraction of a percent this indicator seems to be unreliable. A better idea might be to monitor the cpu-load directly in the callback. (but currently no idea how to accomplish that).

```python
q = queue.Queue()

duration = 10.0 # seconds
nr_samples = int(duration * fs)

nr_buffers = 5
nr_total_runs = 10
# preallocation of memory
audio_buffer = [np.zeros(nr_samples, dtype=np.float32) for k in range(nr_buffers)]

# samples in each buffer
nr_samples_buffer = np.zeros(nr_buffers, dtype=np.uint32)

def callback_ref3(q, indata, frames, time, status):
    if status:
        print(status)
    q.put(indata.copy())
    
wrapped_cb3 = partial(callback_ref3, q)
    
print("begin capturing data")
idx = 0
count_samples = 0
buffer_id = 0
nr_runs = 0
# the processing of audio data is done in the while loop thus freeing resources 
# from the callback function
with sd.InputStream(samplerate=fs, device=1, channels=1, callback=wrapped_cb3) as inp:
    while True:
        data = q.get()
        ndata = len(data)
        count_samples += ndata
        if count_samples >= nr_samples:
            count_samples = ndata
            idx = 0
            nr_runs += 1
            buffer_id = (buffer_id + 1) % nr_buffers
            print(f"cpu_load: {inp.cpu_load}")
        # put data into current buffer    
        audio_buffer[buffer_id][idx:idx + ndata] = data[:,0]
        idx = idx + ndata
        nr_samples_buffer[buffer_id] = idx
        
        if nr_runs >= nr_total_runs:
            break
        
print("end capturing data")
```

```python
# the number of samples in each buffer ...
nr_samples_buffer
```

```python
fig4 = plt.figure(4, figsize=[8, 8])
ax_f4 = fig4.subplots(nr_buffers, 1)
for k in range(nr_buffers):
    ax_f4[k].plot(audio_buffer[k][:nr_samples_buffer[k]], linewidth=0.5, alpha=0.5,  color='k', label=f'buffer: {k}')
    ax_f4[k].grid(True)
    ax_f4[k].set_ylabel('y')
    ax_f4[k].legend()
    
ax_f4[0].set_title(f"audio recording: fs = {fs:10.2f} [Hz]")
ax_f4[-1].set_xlabel('[n]')
```

## Saving Audio Samples to a File

Several options are available.

The most direct approach would be to save audio samples stored in one / many `ndarrays` into file. 

Using the python module `soundfile` permits to store not only a concatenation of ndarrays but *extra information* like sampling rate, number of channels, audio file type.

So this should be the preferred approach.

Below some example code shows how to do it:

1) an audio file is opened for writing in to a file of type `wav`

2) the sampling rate and the number of channels are provided

3) Two `ndarrays` are written

4) Finally the audio file is closed

```python
sfid = sf.SoundFile('./recordings/multi_buffer.wav', mode='w', samplerate=fs, channels=1)
sfid.write(audio_buffer[4][:nr_samples_buffer[4]])
sfid.write(audio_buffer[2][:nr_samples_buffer[2]])
sfid.close()
```
